## Matplotlib Exercises

1. Use matplotlib to plot the following equation:

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = range(-100, 100)
y = [x ** 2 - x + 2 for x in x]

plt.plot(x,y, c = "orange")
plt.annotate('(0,0)', xy=(0, 0), xytext=(50, 40),
             arrowprops={'facecolor': 'green'})
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
x = range(0,100)
y = [math.sqrt(i) for i in x]
plt.plot(x,y)
plt.show()



x = np.arange(-10,10,1)
y = [i ** 3 for i in x]
plt.plot(x,y, c="green")
plt.show()



x = np.arange(-80,80, .1)
y = [math.tan(i) for i in x]
plt.plot(x,y, c="blue")
plt.show()



x = np.arange(0, 30)
y = [2 ** i for i in x]
plt.plot(x,y, c="red")
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python

plt.figure(figsize = (12,10))
plt.suptitle('My Title', color = 'red', fontsize = 20)


plt.subplot(2,2,1)
x = range(0,100, 1)
y = [math.sqrt(i) for i in x]
plt.plot(x,y)


plt.subplot(2,2,2)
x = np.arange(-10,10,1)
y = [i ** 3 for i in x]
plt.plot(x,y, c="green")


plt.subplot(2,2,3)
x = np.arange(-80,80, .1)
y = [math.tan(i) for i in x]
plt.plot(x,y, c="blue")


plt.subplot(2,2,4)
x = np.arange(0, 30)
y = [2 ** i for i in x]
plt.plot(x,y, c="red")
```
