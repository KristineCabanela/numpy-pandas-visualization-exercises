```python
import matplotlib.pyplot as plt
import math
import random
import numpy as np
```

```python

plt.figure(figsize = (10,8))



x = range(1, 20)

# O(1)
y1 = [1 for n in x]

# O(log n)
y2 = [math.log(n) for n in x]

# O(n)
y3 = x

# O(n log n)
y4 = [n * math.log(n) for n in x]

# O(n^2)
y5 = [n**2 for n in x]

# O(2^n)
y6 = [2**n for n in x]



# O(n!)
y7 = [math.factorial(n) for n in x]

# O(n^n)
y8 = [n**n for n in x]






plt.figure(figsize = (16, 8))

plt.title('Big O Notation', fontsize = 20)

plt.xlabel('Elements', fontsize=14)
plt.ylabel('Operations', fontsize=14)

# O(1)
plt.plot(x, y1, label='$\mathcal{O}(1)$', color='red')

# lightcoral, orange, mediumseagreen, cadetblue, dodgerblue, slateblue, violet, crimson

# O(log n)
plt.plot(x, y2, label='$\mathcal{O}(logn)$', color='blue')

# O(n)
plt.plot(x, y3, label='$\mathcal{O}(n)$', color='pink')

# O(n log n)
plt.plot(x, y4, label='$\mathcal{O}(nlogn)$', color='orange')

# O(n^{2})
plt.plot(x, y5, label='$\mathcal{O}(n^{2})$', color='teal')

# O(2^{n})
plt.plot(x, y6, label='$\mathcal{O}(2^{n})$', color='black')

# O(n!)
plt.plot(x, y7, label='$\mathcal{O}(n!)$', color='red')

# O(n^{n})
plt.plot(x, y8, label='$\mathcal{O}(n^{n})$', color='yellow')

plt.xlim(1, 11)
plt.ylim(0, 50)



plt.legend(loc='upper right', fontsize=14)

plt.show()
```
